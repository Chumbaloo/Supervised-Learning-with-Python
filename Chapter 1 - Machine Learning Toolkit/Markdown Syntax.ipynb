# Markdown Syntax Summary

The structure and level of headings is based on the number of # at the start of the line:

# Heading Level 1
## Heading Level 2
### Heading Level 3
#### Heading Level 4


## Lists
We can also do:

* Bullet
* Point
* Lists, using an asterix followed by a space starting
* each line


For enumerated lists:

1. Simply
2. Start each
3. Line with the number and a full stop 
4. After the number

## Formatting

We can do **bold text** using two ** before and after the text we would like to be bold

We can also add *italics* using one * before and after the text to italicise.

> We can do blockquotes

A horizontal rule

-----------------------

Code highlighting
(Python)

```python
str = "code highlighting"
print(str)
```
With different languages
(Bash)

```bash
export STR="code highlighting"
echo ${STR}
```

## Images

We can include images:
![packt_image](packt.png)


## Hyperlinks

And [links](https://packtpub.com)


## HTML

We can also include HTML

<div style="background-color: lightblue;">
<table>
    <tr><th>Col A</th><th>Col B</th></tr>
    <tr><td>1</td><td>2</td></tr>
</table>
</div>
